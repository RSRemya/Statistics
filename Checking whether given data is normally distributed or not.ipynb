```python
#import the dataset
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/shrikant-temburwar/Wine-Quality-Dataset/master/winequality-red.csv",sep=';')
data.head()
```

```python
import seaborn as sns
sns.distplot(data['volatile acidity'],kde=True)
```

```python
sns.distplot(data['citric acid'],kde=True,color='red')
```

```python
# To check whether above distribution are normal or not use: Shapiro Wilk Test
# If p value is less than 0.05 --> Distribution is not normal
```

```python
from scipy import stats
stats.shapiro(data['volatile acidity'])
```

```python
stats.shapiro(data['citric acid'])
```

```python
# Since p-value of both volatile acidity and citric acid is less than 0.05  ----> Distribution are not normal
```

```python
#Another test for checking whether dist is normal or not is :  Normal test
# here also if p value is less than or equal to 0.05 --> Distribution is not normal
stats.normaltest(data['citric acid'])
```

```python
# Making a normally distributed data having 500 data points
from numpy.random import randn
mydata=randn(500)
```

```python
import seaborn as sns
sns.distplot(mydata)
```

```python
from scipy import stats
result = stats.shapiro(mydata)
alpha=0.05
# Null Hypothesis : data was drawn from a normal distribution.
```

```python
if alpha>=result[1]:
    print("Reject the null hyphothesis")
else:
    print("Accept the null hyphothsis")
```
